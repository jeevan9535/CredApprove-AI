```python
import pandas as pd

df = pd.read_csv("final_dataset.csv")

df.head()
```

```python
df.shape
```

```python
df['Approval_Status'] = df['TARGET'].map({
    0: "Approved",
    1: "Risky"
})
```

```python
df['Approval_Status'].value_counts()
```

```python
df['AGE'] = (df['DAYS_BIRTH'] / -365).astype(int)
```

```python
df[['AGE','DAYS_BIRTH']].head()
```

```python
df['EMPLOYMENT_YEARS'] = (df['DAYS_EMPLOYED'] / -365).astype(int)
```

```python
dashboard_df = df[[
    'AGE',
    'AMT_INCOME_TOTAL',
    'NAME_INCOME_TYPE',
    'NAME_EDUCATION_TYPE',
    'NAME_FAMILY_STATUS',
    'NAME_HOUSING_TYPE',
    'CNT_CHILDREN',
    'CNT_FAM_MEMBERS',
    'FLAG_OWN_CAR',
    'FLAG_OWN_REALTY',
    'OCCUPATION_TYPE',
    'EMPLOYMENT_YEARS',
    'Approval_Status'
]]
```

```python
dashboard_df.head()
```

```python
dashboard_df.to_csv("credit_risk_dashboard_dataset.csv", index=False)
```
